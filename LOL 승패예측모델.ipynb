```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

```python
train_index = sorted(np.random.choice(dat_cl.index, size=len(dat_cl.index)*9//10, replace=False))
test_index = np.setdiff1d(dat_cl.index, train_index)
```

```python
train_dataset = data_parse_frame_r.iloc[train_index, :]
```

```python
test_dataset = data_parse_frame_r.iloc[test_index, :]
```

```python
def build_model():
  model = keras.Sequential([
    layers.Dense(4, activation='relu', input_shape=[len(data_parse_frame_r.columns)]),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
  ])

  optimizer = tf.keras.optimizers.RMSprop(0.001)

  model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'accuracy'])
  return model
```

```python
model = build_model()
```

```python
model.summary()
```

```python
history = model.fit(train_dataset, train_result, epochs=30, validation_data=(test_dataset, test_result)  )
```
